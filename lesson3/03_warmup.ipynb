````python
"""
This is a function which plots both the Empirical CDF and Analytical CDF and then compare them in one plot.
"""

import numpy as np
import scipy.special
import matplotlib.pyplot as plt

N = 1000 # select 1000 points
np.random.seed(42) 
x = np.random.normal(size = N) 

sorted_x = np.sort(x)

ecdf_x = np.arange(1, N+1) / N # calculation of the empirical CDFs.

def normal_cdf(x): # a function to calculate the analytical solution of CDFs.
    """ Calculates the CDF of the standard normal distribution at x."""
    return (1 + scipy.special.erf(x / np.sqrt(2))) / 2

plt.plot(sorted_x, ecdf_x, label = 'Empirical CDF')
plt.plot(sorted_x, normal_cdf(sorted_x), label = 'Analytical CDF')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.legend()
plt.show()

````
